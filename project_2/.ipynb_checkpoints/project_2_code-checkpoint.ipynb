```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import patsy

from sklearn import metrics
from sklearn.linear_model import Ridge, Lasso, ElasticNet, LinearRegression, RidgeCV, LassoCV, ElasticNetCV
from sklearn.model_selection import train_test_split,cross_val_score, KFold

%config InlineBackend.figure_format = 'retina'
%matplotlib inline

plt.style.use('fivethirtyeight')
```

```python
train_df = pd.read_csv('./datasets/train.csv')    # shape is (2051, 81)
test_df = pd.read_csv('./datasets/test.csv')      # shape is (879, 80)

col_string = pd.read_csv('./datasets/string_columns.txt',header=None)   
# string_columns.txt is a file that stores all columns who are supposed to be string according to data description
```

```python
# Convert all columns in col_string to string (according to data dictionary)
for i in train_df.columns:
    if i in list(col_string[0]):
        print(i)
        train_df[i] = train_df[i].astype(str)
```
