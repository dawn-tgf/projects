```python
# Create a function to convert ordinal variable to numeric
def ordinal_to_num_col(col,map):
    col_name = col.lower().replace(' ','_') + "_num"
    test_df[col_name] = test_df[col].astype(str).map(lambda x: map[x])
    print(col_name, "created")
```

### Engineer Numeric cols from Numeric Cols

```python
# Create numeric cols from numeric cols
new_num_cols = ['artery_st','feeder_st','railroad','park','age',\
                'age_remod', 'total_baths','hse_cond','garage_cond',]

test_df['age'] = 2020 - test_df['Year Built']
test_df['age_remod'] = 2020 - test_df['Year Remod/Add']
test_df['total_baths'] = test_df['Full Bath'] + 0.5* test_df['Half Bath']
test_df['bsmt_fin_sf'] = test_df['BsmtFin SF 1'] + test_df['BsmtFin SF 2']

test_df['hse_cond'] = round( test_df['Overall Qual']/test_df['Overall Qual'].mean() \
                             + test_df['Overall Cond']/test_df['Overall Cond'].mean()\
                             + test_df['exter_qual_num']/test_df['exter_qual_num'].mean()\
                             + test_df['exter_cond_num']/test_df['exter_cond_num'].mean() \
                             + test_df['functional_num']/test_df['functional_num'].mean() \
                             ,2)
test_df['garage_cond'] = round( test_df['garage_qual_num']/test_df['garage_qual_num'].mean()\
                                + test_df['garage_cond_num']/test_df['garage_cond_num'].mean()  \
                                + test_df['garage_finish_num']/test_df['garage_finish_num'].mean()\
                                ,2)
test_df['bsmt_gen_cond'] = round( test_df['bsmt_cond_num']/test_df['bsmt_cond_num'].mean()\
                                  + test_df['bsmt_qual_num']/test_df['bsmt_qual_num'].mean()  \
                                  + test_df['bsmtfin_type_1_num']/test_df['bsmtfin_type_1_num'].mean()\
                                  + test_df['bsmtfin_type_2_num']/test_df['bsmtfin_type_2_num'].mean()
                                ,2)
test_df.head()
```

### Engineer the Norminal cols

```python
# From 'Condition 1' and 'Condition 2'
# create dummy cols for artery_st(artery) , feeder_st(feedr), railroad(RRAn,RRAe,RRNn,RRNe), park (PosN,PosA)
# for any feature: if a loc has the same features in both cols, it'll be given higher weightage (2)

# create col for 'artery_st'
test_df['artery_st'] = (test_df['Condition 1'] =='Artery').astype(int) + (test_df['Condition 2'] =='Artery').astype(int)

# create col for 'feeder_st'
test_df['feeder_st'] = (test_df['Condition 1'] =='Feedr').astype(int) + (test_df['Condition 2'] =='Feedr').astype(int)

# Create col for 'railroad'
test_df['railroad'] = (test_df['Condition 1'].isin(['RRAn','RRAe','RRNn','RRNe'])).astype(int) \
                     + (test_df['Condition 2'].isin(['RRAn','RRAe','RRNn','RRNe'])).astype(int)

# Create col for 'park'('PosN','PosA')
test_df['park'] = (test_df['Condition 1'].isin(['PosN','PosA'])).astype(int) \
                 + (test_df['Condition 2'].isin(['PosN','PosA'])).astype(int)
```
